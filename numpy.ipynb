SLicing

```python
import numpy as np
a=[1,-2,4,6.2,'abc',True]
b=np.array(a)
print("1:",a[2:4])
print("2:",a[3:7])
print("3:",a[0:7])
print("4:",a[ :4])
print("5:",a[0:])
print("6:",a[ : ])
print("7:",a[2:7:2])
print("8:",a[-4:-2])
print("9:",a[-2:-4])
print("10:",a[4:2])
print("11:",a[-4:-2:-2])



```

Zero's method

```python
import numpy as np
a=np.zeros((4,3))
print(a)
b=np.zeros(4)
print(b)
c=np.zeros(4,dtype=int)
print(c)
d=np.zeros((4,2),dtype=int)
print(d)
```

One's method

```python
import numpy as np
a=np.ones((2,3))
print(a)
b=np.ones(2)
print(b)
c=np.ones((2,3),dtype=int)
print(c)
```

Eye method

```python
import numpy as py
a=py.eye(1,5,3)
print(a)
b=py.eye(5,5,1)
print(b)
c=py.eye(5,5,4)
print(c)
```
