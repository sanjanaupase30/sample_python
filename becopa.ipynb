Break,pass,continue statements

```python
for i in range(1,10):
    if i==4:
        break
    print(i)
print("**********************************")
for i in range(1,10):
    if i==4:
       pass
    print(i)

print("**********************************")
for i in range(1,10):
    if i==4:
       continue
    print(i)





 
```
