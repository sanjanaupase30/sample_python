for loop

```python
name="sanjana"
for i in name:
    print(i)
print("_____________________________________")
numbers=[1,2,3,4,5,6]
for i in numbers:
    print(i)
print("_____________________________________")
for i in range(1,20):
    print(i)
print("_____________________________________")
for i in range(1,20):
    i=i+2
    print(i)
print("_____________________________________")
for i in range(1,5):
    for j in range(1,10):
        print(i,j)

print("_____________________________________")
for i in range(1,5):
    if i%2==0:
        print(i,"even")

    
```
